**EXERCICÍO 5**

```python
nome = input("Digite seu nome: ")

# while para validar as informações

# <= pra que se for menor ou igual a 3 invalido
while len(nome) <= 3:
    print("Nome inválido.")
    nome = input("Digite seu nome novamente: ")

idade = int(input("Digite sua idade: "))

# or pra que se a idade for menor que zero ou maior que 150 não for valida
while idade < 0 or idade > 150:
    print("Idade inválida.")
    idade = int(input("Digite sua idade novamente: "))

salario = float(input("Digite seu salário: "))

# pra que se for menor ou igual a 0 invalido
while salario <= 0:
    print("Salário inválido.")
    salario  = float(input("Digite seu salário novamente: "))

sexo = input("Digite seu sexo (Feminino ou Masculino): ")

# != para ver quando a pessoa digitou algo diferente do que era pedido
while sexo != "Feminino" and sexo != "Masculino":
    print("Sexo inválido.")
    sexo = input("Digite seu sexo novamente: ")

estado_civil = input("Digite seu estado civil (Solteiro, Casado, Viuvo, Divorciado): ")

while estado_civil != "Solteiro" and estado_civil != "Casado" and estado_civil != "Viuvo" and estado_civil != "Divorciado":
    print("Estado civil inválido.")
    estado_civil = input("Digite seu estado civil novamente: ")

print("Dados cadastrados com sucesso!")
```

**EXERCICÍO 6**

```python
numero = int(input("Digite um número inteiro: "))

primo = True

# o 1 não é considerado número primo
if numero <= 1:
    primo = False

# alternei entre for, if, e else
for i in range(2, numero):

    # usei o % para ver se o resto da divisão é 0
    if numero % i == 0:
        primo = False

if primo:
    print("Esse número é primo")
else:
    print("Esse número não é primo")
```

**EXERCICÍO 7**

**EXERCICÍO 8**

```python
lista = [5, 7, 2, 9, 4, 1, 3]

# len pra saber o tamanho da lista
print("Tamanho da lista:")
print(len(lista))


# max variavel pra saber o maior número da lista
print("Maior número:")
print(max(lista))


# min variavel pra saber o menor número da lista
print("Menor número:")
print(min(lista))


# sum pra somar todos os números da lista juntos e dar o valor
print("Soma dos números:")
print(sum(lista))


# sort organiza a lista em ordem crescente
print("Ordem crescente:")
lista.sort()
print(lista)


# sort + reverse=true para organizar em ordem crescente e reverter na ordem decrescente
print("Ordem decrescente:")
lista.sort(reverse=True)
print(lista)
```
